```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings
```

```python
df = pd.read_csv('Scale_data.txt', header=0)
```

```python
df
```

```python
df.dtypes
df['Scale(m)'] = df['Scale(m)'].astype('float64')
```

```python
df
```

```python
df.columns = ['Incendio', 'Scale', 'Size']
```

```python
df
```

```python
df['Scale_factor'] = df['Scale'] / df['Size']
df
```

```python
df.describe()
```

```python
df['Scale_Ref_min'] = 1 / df['Scale_factor'] * df['Scale_factor'].min()
df['Scale_Ref_median'] = 1 / df['Scale_factor'] * df['Scale_factor'].median()
df['Scale_Ref_max'] = 1 / df['Scale_factor'] * df['Scale_factor'].max()
```

```python
df
```

```python
df.to_csv('scale_data.csv', index=False)
```

```python
import os
import fitz

# Establecer el directorio de origen
directory = ".\pdf"

# Iterar a través de todos los archivos PDF en el directorio
for filename in os.listdir(directory):
    if filename.endswith(".pdf"):
        # Abrir archivo PDF
        filepath = os.path.join(directory, filename)
        doc = fitz.open(filepath)
        # Obtener la primera página (en este caso, se asume que cada archivo PDF tiene una sola página)
        page = doc[0]
        # Crear una imagen PNG de la página con una resolución de 300 ppp
        pix = page.get_pixmap()
        # Guardar la imagen PNG en el mismo directorio con el mismo nombre y extensión .png
        pix.save(os.path.splitext(filepath)[0] + ".png")
```
